## K-Means

```python
def kmeans_range(k_range, y, name):

    k_to_results = {}

    print(f"<=== [{name}] ===>")

    for k in k_range:

        print(f"K-Means with k = {k}", end="")

        t0 = time()
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=SEED)
        kmeans.fit(y)

        silhoutte_score = silhouette_score(y[::10], kmeans.labels_[::10])
        k_to_results[k] = (kmeans, silhoutte_score)

        t1 = time()

        print(f" -> Done! [{round(t1 - t0, 2)} s]")
        print(f"Silhouette Score = {silhoutte_score:.5f}, WSS = {kmeans.inertia_}\n")

    return k_to_results
```

```python
k_to_results_TFIDF = kmeans_range(range(2, 21), X_TFIDF, 'TF-IDF')
k_to_results_ST = kmeans_range(range(2, 21), X_ST, 'ST')
k_to_results_ST_FT = kmeans_range(range(2, 21), X_ST_FT, 'ST_FT')
```

Let's compute the **medoids** for each cluster in each setting:

```python
def compute_medoid(data_2d, labels):
    medoid_indices = []
    for label in np.unique(labels):
        cluster_mask = labels == label
        cluster_points = data_2d[cluster_mask]
        # Pairwise distances within cluster
        distances = np.sum(np.linalg.norm(cluster_points[:, None] - cluster_points[None, :], axis=2), axis=1)
        # Index relative to the whole dataset
        medoid_index = np.where(cluster_mask)[0][np.argmin(distances)]
        medoid_indices.append(medoid_index)
    medoids = data_2d[medoid_indices]
    return medoids
```

```python
medoids_TFIDF = compute_medoid(umap_TFIDF, label_kmeans_TFIDF)
medoids_ST = compute_medoid(umap_ST, label_kmeans_ST)
medoids_ST_FT = compute_medoid(umap_ST_FT, label_kmeans_ST_FT)
```

```python
from sklearn.metrics.pairwise import cosine_distances
from sklearn.preprocessing import normalize
```

```python
def fit_dbscan(X, name, eps=None, min_samples=None):

    print(f"<=== [{name}] ===>")

    t0 = time()
    if eps is not None and min_samples is not None:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    else:
        dbscan = DBSCAN()

    dbscan.fit(X)
    unique_labels, counts = np.unique(dbscan.labels_, return_counts=True)

    print(f"Number of clusters: {len(unique_labels)-1}")
    print(f"Number of noise points: {counts[0]}")
    print(f"Average cluster size: {np.mean(counts[1:])}")
    print(f"Std cluster size: {np.std(counts[1:])}")

    silhoutte_score = silhouette_score(X[::10], dbscan.labels_[::10])
    print(f"Silhouette Score = {silhoutte_score:.5f}\n")

    t1 = time()

    return dbscan# , silhoutte_score
```

# 3.0 Topic Modeling

After clustering, we perform **topic modeling** to the same subset of data (transformed in the same way). Thus, unlike before, we are able to express the data points as mixtures of non-exclusive clusters.
For this task, we are going to employ **LDA** (Latent Dirichlet Allocation).

First, we split the data into a train and test set:

```python
documents_train, documents_test = train_test_split(X, test_size=0.4)
print('No. training docs:', len(documents_train))
print('No. testing docs:', len(documents_test))
```

LDA is a probabilistic models which reasons on top of raw term counts (**BoW**), hence we are going to feed it documents encoded by `CountVectorizer`.

```python
vectorizer = CountVectorizer(stop_words='english', min_df=5, max_df=0.5)
vector_documents_train = vectorizer.fit_transform(documents_train)
```

We select `n_components = 30` to match the number of components previously selected in the *Clustering* section for comparison and iterate the training loop:

```python
lda = LatentDirichletAllocation(n_components=30, verbose=1, learning_method='online', max_iter=30)
lda.fit(vector_documents_train)
```

Let's compute the perplexity on the test set to have an idea of the performance:

```python
vector_documents_test = vectorizer.transform(documents_test)
lda.perplexity(vector_documents_test)
```

For each topic, let's compute the most important words for each topic:

```python
for i in range(len(lda.components_)):
    sorted_terms = lda.components_[i].argsort()[::-1]
    concatenated_terms = '[' + ', '.join(vocab[i] for i in sorted_terms[:10]) + ']'
    print (f'Topic {i + 1}:\t', concatenated_terms)
```

The algorithm finds some insightful info! E.g.:
- Topic n. 1 seems to aggregate terms semantically related to **statistics** ("analysis", "variance", "linear", ...).
- Topic n. 6 seems to aggregate terms related to **children** ("child", "infant", "newborn", ...).
- Topic n. 16 seems to aggregate terms related to **weight** ("body", "mass", "obesity", ...).

However, a more careful selection of the number of components might refine the results, since the topics tend to have some overlapping.
Anyway, these results should be validated by experts of the field.

Moving on, each term is associated with a certain **importance** w.r.t. each topic. Let's plot, for every topic, the most important words (the ones just presented) along with their prevalence:

```python
num_words = 10
cols = 5
rows = int(len(lda.components_)/5)

fig, axes = plt.subplots(rows, cols, figsize=(14, 4 * rows), sharex=True)
axes = axes.flatten()
for topic_idx, topic in enumerate(lda.components_):
    top_terms_index = topic.argsort()[:-num_words - 1:-1]
    top_terms = [vocab[i] for i in top_terms_index]
    weights = topic[top_terms_index]
    ax = axes[topic_idx]
    ax.barh(top_terms, weights, height=0.7)
    ax.set_title(f'Topic {topic_idx +1}',fontdict={'fontsize': 15})
    ax.invert_yaxis()
    ax.set_xscale('log')

plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
plt.tight_layout()
plt.show()
```
